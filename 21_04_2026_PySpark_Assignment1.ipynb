```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Pyspark Exercise 1").getOrCreate()
data=[("Karl",340000,"IT"),
      ("Daryl",50000,"Management"),
      ("Jim",30000,"Sales"),
      ("Kenneth",50000,"HR"),
      ("Sam",450000,"IT")]
columns=["Name","Salary","Department"]
df=spark.createDataFrame(data,columns)
df.show()
```

```python
df.select("Name","Salary").show()
```

```python
df.filter(df.Salary>50000).show()
```

```python
df.groupBy("Department").count().show()
```

```python
df.sort("Salary").show()
```
