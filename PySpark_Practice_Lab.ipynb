```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.window import Window

spark = SparkSession.builder.appName("SQL_Pyspark_Exercise").getOrCreate()
sc = spark.sparkContext
customers = spark.read.csv("customers.csv", header=True, inferSchema=True)
products = spark.read.csv("products.csv", header=True, inferSchema=True)
orders = spark.read.csv("orders.csv", header=True, inferSchema=True)
order_items = spark.read.csv("order_items.csv", header=True, inferSchema=True)
payments = spark.read.csv("payments.csv", header=True, inferSchema=True)
```

```python
#DATAFRAME EXERCISE
customers.show()
customers.printSchema()
customers.count()
customers.show(5)
customers.select("name","city").show()
products.count()
products.select("product_name","price").show()
orders.show()
orders.count()

payments.show()
```

```python
customers.filter(customers.city == "Hyderabad").show()

customers.filter(customers.age > 30).show()

products.filter(products.price > 10000).show()

products.filter(products.category == "Electronics").show()

orders.filter(orders.status == "Delivered").show()

orders.filter(orders.status == "Cancelled").show()

payments.filter(payments.payment_type == "UPI").show()

customers.filter(customers.city.isin("Bangalore","Delhi")).show()

products.filter(products.price < 1000).show()

customers.filter((customers.age >= 25) & (customers.age <= 35)).show()
```

```python
orders.join(customers, "customer_id").show()

orders.join(customers, "customer_id") \
.select("name","status").show()

orders.join(order_items, "order_id").show()

order_items.join(products, "product_id").show()

df = order_items.join(products, "product_id") \
.withColumn("revenue", col("quantity") * col("price"))

df.show()

full = orders.join(customers,"customer_id") \
.join(order_items,"order_id") \
.join(products,"product_id")

full.select("name","product_name","quantity").show()

full.withColumn("revenue", col("quantity") * col("price")) \
.groupBy("order_id").sum("revenue").show()

full.withColumn("revenue", col("quantity") * col("price")) \
.groupBy("product_name").sum("revenue").show()

full.withColumn("revenue", col("quantity") * col("price")) \
.groupBy("name").sum("revenue").show()
```

```python
orders.groupBy("customer_id").count().show()

orders.join(customers,"customer_id") \
.groupBy("city").count().show()

rev = full.withColumn("revenue", col("quantity") * col("price"))

rev.groupBy("product_name").sum("revenue").show()

rev.groupBy("category").sum("revenue").show()

rev.groupBy("city").sum("revenue").show()

rev.groupBy("product_name").sum("quantity").show()

rev.groupBy("category").sum("quantity").show()

rev.groupBy("name").sum("revenue").show()

rev.groupBy("name").sum("revenue") \
.orderBy(col("sum(revenue)").desc()).show(5)

rev.groupBy("product_name").sum("revenue") \
.orderBy(col("sum(revenue)").desc()).show(3)
```

```python
w = Window.orderBy(col("price").desc())
products.withColumn("rank", rank().over(w)).show()

w2 = Window.partitionBy("category").orderBy(col("price").desc())
products.withColumn("rank", rank().over(w2)).show()

w3 = Window.orderBy("signup_date")
customers.withColumn("row_num", row_number().over(w3)).show()

rev2 = rev.groupBy("name").sum("revenue")

w4 = Window.orderBy(col("sum(revenue)").desc())
rev2.withColumn("rank", rank().over(w4)).show()

rev2.withColumn("dense_rank", dense_rank().over(w4)).show()

w5 = Window.partitionBy("city").orderBy(col("sum(revenue)").desc())
rev.groupBy("city","name").sum("revenue") \
.withColumn("rank", rank().over(w5)).filter(col("rank")==1).show()

w6 = Window.partitionBy("category").orderBy(col("revenue").desc())
rev.withColumn("rank", rank().over(w6)).filter(col("rank")==1).show()

rev.withColumn("running_total",
    sum("revenue").over(Window.orderBy("order_id"))
).show()

rev.withColumn("running_total",
    sum("revenue").over(Window.partitionBy("product_name").orderBy("order_id"))
).show()

payments.withColumn("rank",
    rank().over(Window.orderBy(col("amount").desc()))
).show()
```

```python
customers.createOrReplaceTempView("customers")
orders.createOrReplaceTempView("orders")
products.createOrReplaceTempView("products")
payments.createOrReplaceTempView("payments")
order_items.createOrReplaceTempView("order_items")
```

```python
spark.sql("SELECT * FROM customers").show()

spark.sql("SELECT * FROM customers WHERE age > 30").show()

spark.sql("SELECT city, COUNT(*) FROM customers GROUP BY city").show()

spark.sql("SELECT city, AVG(age) FROM customers GROUP BY city").show()

spark.sql("""
SELECT c.name, o.order_id
FROM customers c JOIN orders o
ON c.customer_id = o.customer_id
""").show()

spark.sql("SELECT customer_id, COUNT(*) FROM orders GROUP BY customer_id").show()

spark.sql("""
SELECT * FROM order_items oi
JOIN products p ON oi.product_id = p.product_id
""").show()

spark.sql("""
SELECT oi.order_id, SUM(oi.quantity * p.price) AS revenue
FROM order_items oi JOIN products p
ON oi.product_id = p.product_id
GROUP BY oi.order_id
""").show()

spark.sql("""
SELECT c.name, SUM(oi.quantity * p.price) AS revenue
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
GROUP BY c.name
ORDER BY revenue DESC
LIMIT 3
""").show()

spark.sql("""
SELECT p.product_name, SUM(oi.quantity * p.price) AS revenue
FROM order_items oi JOIN products p
ON oi.product_id = p.product_id
GROUP BY p.product_name
ORDER BY revenue DESC
LIMIT 1
""").show()
```
