```python
df.filter(df["salary"]>90000).show()
df.filter(df["department"]=="Engineering").show()
df.filter(df["department"]=="Marketing").show()
df.filter(df["experience"]>6).show()
df.filter(df["salary"]<70000).show()
df.filter(df["city"]=="London").show()
df.filter((df["salary"]>90000) & (df["department"]=="Finance" )).show()
df.filter((df["experience"]>5) & (df["department"]=="Engineering" )).show()
df.filter((df["salary"]>85000) & (df["salary"]<=90000 )).show()
df.filter(df["experience"]<4).show()
```

```python
df=df.withColumn("Annual Salary",df["salary"]*12)
df.show()
df=df.withColumn("Salary in thousands",df["salary"]/1000)
df.show()
df = df.withColumn("experience_double",df["experience"]*2)
df.show()
df = df.withColumn("company", lit("GlobalTech"))
df.show()
df = df.withColumn("salary after bonus",df["salary"]+5000)
df.show()
df = df.withColumn("salary after tax",df["salary"]*0.9)
df.show()
df = df.withColumn("experience next year",df["experience"]+1)
df.show()
df = df.withColumn("salary per year of experience",df["salary"]/df["experience"])
df.show()
df = df.withColumnRenamed("salary","monthly salary")
df.show()
df = df.drop("experience double")
df.show()
```

```python
from pyspark.sql.functions import sum, avg, max, min, count
df.select(sum("monthly salary")).show()
df.select(avg("monthly salary")).show()
df.select(max("monthly salary")).show()
df.select(min("monthly salary")).show()
df.select(count("*")).show()
df.groupBy("department").count().show()
df.groupBy("department").avg("monthly salary").show()
df.groupBy("department").max("monthly salary").show()
df.groupBy("department").min("monthly salary").show()
df.groupBy("department").sum("monthly salary").show()
```
