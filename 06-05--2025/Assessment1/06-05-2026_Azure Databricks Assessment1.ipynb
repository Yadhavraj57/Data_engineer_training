```python
from pyspark.sql.functions import *

patients_df = patients_df.withColumn("registration_date", to_date("registration_date"))
visits_df = visits_df.withColumn("visit_date", to_date("visit_date"))

visits_df = visits_df.withColumn("test_cost", visits_df["tests_count"] * 500)

visits_df = visits_df.withColumn("estimated_total_bill", visits_df["test_cost"] + 1000)

patients_df = patients_df.withColumn("age_category",when(patients_df["age"] < 30, "Young").when(patients_df["age"] < 50, "Adult").otherwise("Senior"))

patients_df = patients_df.withColumn("membership_priority",when(patients_df["membership"] == "Gold", 1).when(patients_df["membership"] == "Silver", 2).otherwise(3))

visits_df = visits_df.withColumn("visit_priority",when(visits_df["visit_status"] == "Completed", 1).when(visits_df["visit_status"] == "Pending", 2).otherwise(3))

visits_df = visits_df.withColumn("high_bill_flag",when(visits_df["estimated_total_bill"] > 5000, "Yes").otherwise("No"))

patients_df = patients_df.withColumnRenamed("patient_name", "full_name")
```

```python

pv = patients_df.join(visits_df, "patient_id")
pvd = pv.join(doctors_df, "doctor_id")
final_df = pvd.join(payments_df, "visit_id")

final_df.select("full_name","city","doctor_name","specialization","visit_status","bill_amount").show()

final_df.filter(col("city") != col("doctor_city")).show()

final_df.filter("visit_status='Completed' AND payment_status='Paid'").show()
final_df.filter("visit_status='Pending' AND payment_status='Pending'").show()
final_df.filter("visit_status='Cancelled' AND payment_status='Cancelled'").show()

final_df.groupBy("patient_id").count().filter("count > 1").show()
```
