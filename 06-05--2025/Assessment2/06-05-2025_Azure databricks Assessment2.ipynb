```python
from pyspark.sql.functions import *

orders_df = orders_df.withColumn("order_date", to_date("order_date"))
orders_df = orders_df.withColumn("total_order_value", orders_df["quantity"] * 1000)
products_df = products_df.withColumn("stock_status",when(products_df["stock_quantity"] < 20, "Low").otherwise("Sufficient"))

orders_df = orders_df.withColumn( "order_priority",when(orders_df["order_status"] == "Delivered", 1).when(orders_df["order_status"] == "Pending", 2).otherwise(3))

products_df = products_df.withColumn("expensive_product_flag",when(products_df["price"] > 20000, "Yes").otherwise("No"))

products_df = products_df.withColumn("category", upper("category"))

products_df = products_df.withColumnRenamed("warehouse_city", "inventory_city")

products_df = products_df.withColumn("inventory_value",products_df["price"] * products_df["stock_quantity"])

products_df = products_df.withColumn("low_stock_flag",when(products_df["stock_quantity"] < 20, "Yes").otherwise("No"))
```

```python
final_df = products_df.join(orders_df, "product_id").join(suppliers_df, "supplier_id").join(payments_df, "order_id")
final_df.select("product_name","supplier_name","quantity","bill_amount").show()
final_df.filter("order_status='Delivered' AND payment_status='Paid'").show()
final_df.filter("order_status='Pending' AND payment_status='Pending'").show()
final_df.filter("order_status='Cancelled' AND payment_status='Cancelled'").show()
final_df.groupBy("product_id").count().filter("count>1").show()
```

```python
from pyspark.sql.window import Window
w = Window.partitionBy("category").orderBy(desc("bill_amount"))
final_df.withColumn("rank", rank().over(w)).show()
final_df.withColumn("row_num", row_number().over(w)).show()
w2 = Window.partitionBy("supplier_city").orderBy(desc("bill_amount"))
final_df.withColumn("dense_rank", dense_rank().over(w2)).show()
w3 = Window.orderBy("order_date") 
final_df.withColumn("running_total",sum("bill_amount").over(w3)).show()
```
