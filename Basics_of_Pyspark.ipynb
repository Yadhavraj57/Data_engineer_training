```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("spark").getOrCreate()
sc=spark.sparkContext
print("spark ready")
```

```python
number=[1,2,3,4,5]
rdd=sc.parallelize(number)
print(rdd)
rdd.collect()
```

```python
names=["Yadhav","Priya","Jammy"]
Names_rdd=sc.parallelize(names)
Names_rdd.collect()
```

```python
rdd.getNumPartitions()
rdd=sc.parallelize(number,3)
rdd.collect()
rdd.glom().collect()
```

```python
rddd=sc.parallelize([1,2,3,4,5,6,7],3)

rddd.glom().collect()
```

RDD operations: two categories
1. Transformations
These create a new RDD.
Examples:
map
filter
flatMap
distinct
union
intersection
groupByKey
reduceByKey
2. Actions
These return a result to the driver.
Examples:
collect
count
first
take
reduce

Lasy evaluation:
        The interpreter will save the resources by delaying the transformations
        until an action is explicitly trigerred.
        





```python
from re import X
mapped_rdd=rdd.map(lambda x:X*2)
filtered_rdd=mapped_rdd.filter(lambda x:x>5)
print("Nothing executed yet")

filtered_rdd.collect()
```

```python
squared_rdd=rdd.map(lambda x:x*X)
squared_rdd.collect()
```

```python
upper_rdd=Names_rdd.map(lambda x:x.upper())
upper_rdd.collect()
```

```python

flat_mapped = Names_rdd.flatMap(lambda x: x.split(" "))
flat_mapped.collect()
```

```python
rdd3 = sc.parallelize([1,2,2,3,4,4,5])
rdd3.distinct().collect()
```

```python
rdd_a = sc.parallelize([1,2,3])
rdd_b = sc.parallelize([3,4,5])
rdd_a.union(rdd_b).collect()
```

```python
rdd_a.intersection(rdd_b).collect()
```

```python
pair_rdd = sc.parallelize([("a",1),("b",2),("a",3)])
grouped = pair_rdd.groupByKey()
print([(k, list(v)) for k, v in grouped.collect()])
```

```python
reduced = pair_rdd.reduceByKey(lambda x, y: x + y)
print(reduced.collect())
```

```python
rdd1 = sc.parallelize([1,2,3,4])
rdd2 = sc.parallelize([3,4])

rdd1.subtract(rdd2).collect()
```

```python
rdd2 = sc.parallelize([
    ("Rahul", 85),
    ("Sneha", 92),
    ("Arjun", 78)
])

rdd2.mapValues(lambda x: x + 5).collect()
```

```python
sc.parallelize([10, 40, 20, 30]).sortBy(lambda x: x).collect()
```

```python
lines=sc.parallelize(["python spark python spark hadoop spark hadoop python"])
word_counts=(lines.flatMap(lambda line:line.split(" ")).map(lambda word:(word, 1)).reduceByKey(lambda a, b: a + b))
word_counts.collect()
```
