```sql
SELECT * FROM employees;

SELECT emp_name, city, salary
FROM employees;

SELECT DISTINCT department
FROM employees;

SELECT COUNT(*) AS total_employees
FROM employees;

SELECT *
FROM employees
WHERE city = 'Hyderabad';

SELECT *
FROM employees
WHERE city = 'Bangalore';

SELECT *
FROM employees
WHERE salary > 90000;

SELECT *
FROM employees
WHERE experience > 6;

SELECT *
FROM employees
WHERE department = 'Engineering';

SELECT *
FROM employees
WHERE salary BETWEEN 80000 AND 95000;
```

```sql
SELECT *
FROM employees
ORDER BY salary ASC;

SELECT *
FROM employees
ORDER BY salary DESC;

SELECT *
FROM employees
ORDER BY salary DESC
LIMIT 5;

SELECT *
FROM employees
ORDER BY salary ASC
LIMIT 5;

SELECT *
FROM employees
ORDER BY experience DESC;
```

```sql
SELECT AVG(salary) AS avg_salary
FROM employees;

SELECT MAX(salary) AS max_salary
FROM employees;

SELECT MIN(salary) AS min_salary
FROM employees;

SELECT department, COUNT(*) AS employee_count
FROM employees
GROUP BY department;

SELECT department, AVG(salary) AS avg_salary
FROM employees
GROUP BY department;
```

```sql
SELECT department, SUM(salary) AS total_salary
FROM employees
GROUP BY department;

SELECT department, AVG(experience) AS avg_experience
FROM employees
GROUP BY department;

SELECT city, COUNT(*) AS total_employees
FROM employees
GROUP BY city;

SELECT city, MAX(salary) AS max_salary
FROM employees
GROUP BY city;

SELECT department, MIN(salary) AS min_salary
FROM employees
GROUP BY department;
```

```sql
SELECT *
FROM employees
WHERE salary > (
    SELECT AVG(salary)
    FROM employees
);

SELECT e.*
FROM employees e
JOIN (
    SELECT department, AVG(experience) AS avg_exp
    FROM employees
    GROUP BY department
) d
ON e.department = d.department
WHERE e.experience > d.avg_exp;

SELECT *
FROM employees
WHERE salary < 70000;

SELECT *
FROM employees
WHERE department = 'HR';

SELECT *
FROM employees
WHERE department = 'Finance'
AND salary > 90000;
```

```sql
SELECT e.emp_name, e.department, p.project_name, p.budget
FROM employees e
JOIN projects p
ON e.department = p.department;

SELECT e.emp_name,e.department,p.project_name
FROM employees e
JOIN projects p
ON e.department = p.department;

SELECT department,SUM(budget) AS total_budget
FROM projects
GROUP BY department;

SELECT DISTINCT e.*
FROM employees e
JOIN projects p
ON e.department = p.department;

SELECT p.project_name,COUNT(e.emp_id) AS employee_count
FROM projects p
JOIN employees e
ON p.department = e.department
GROUP BY p.project_name;
```

```sql
CREATE OR REPLACE TABLE employee_salary_analysis AS

SELECT
    emp_name,
    city,
    department,
    salary,
    experience,

    CASE
        WHEN salary >= 95000 THEN 'High'
        WHEN salary >= 80000 THEN 'Medium'
        ELSE 'Low'
    END AS salary_category,

    AVG(salary) OVER (PARTITION BY department) AS department_avg_salary

FROM employees;
```
