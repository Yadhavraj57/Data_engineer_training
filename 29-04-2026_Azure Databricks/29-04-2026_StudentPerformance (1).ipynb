```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg, when, count

spark = SparkSession.builder.appName("StudentPerformance").getOrCreate()

data = [
(101, "Arjun Kumar", "Python", 86, 92),
(102, "Neha Gupta", "Python", 74, 80),
(103, "Rahul Nair", "Data Engineering", 88, 85),
(104, "Priya Sharma", "AI", 91, 95),
(105, "Siddharth Rao", "Data Engineering", 67, 72),
(106, "Ayesha Khan", "AI", 79, 88),
(107, "Kavya Reddy", "Python", 94, 96),
(108, "Manish Yadav", "Data Engineering", 58, 69),
(109, "Divya Menon", "AI", 83, 90),
(110, "Farhan Ali", "Python", 62, 75)
]

columns = ["student_id", "student_name", "course", "marks", "attendance"]

df = spark.createDataFrame(data, columns)
df = df.withColumn("grade",when(df.marks >= 90, "A").when(df.marks >= 75, "B").when(df.marks >= 60, "C").otherwise("Needs Improvement"))
df = df.withColumn("certification_status", when((df.marks >= 75) & (df.attendance >= 80), "Eligible").otherwise("Not Eligible"))
df.groupBy("course").agg(avg("marks").alias("average_marks")).show()
df.filter(df.certification_status == "Eligible").groupBy("course").agg(count("*").alias("eligible_students")).show()
```
